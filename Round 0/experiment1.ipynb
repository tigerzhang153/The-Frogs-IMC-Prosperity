```python
import pandas as pd

df = pd.read_csv('Data (Raw CSV)/prices_round_0_day_-1.csv', sep=';')
```

```python
emeralds = df[df['product'] == 'EMERALDS'].copy()
tomatoes = df[df['product'] == 'TOMATOES'].copy()
```

```python
import matplotlib.pyplot as plt

plt.plot(emeralds['timestamp'], emeralds['mid_price'], label='EMERALDS')
plt.plot(tomatoes['timestamp'], tomatoes['mid_price'], label='TOMATOES')
plt.xlabel('Timestamp')
plt.ylabel('Mid Price')
plt.legend()
plt.title('Mid Price Over Time')
plt.show()
```
